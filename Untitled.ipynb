```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_excel('test_data.xlsx'))
```

```python
df1 =df[df["heuristic steps"]==3].copy()
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes = plt.subplots(figsize = (30,8))
axes.set_xlabel("file_name")
axes.set_ylabel("runtime heuristic steps = 3")
temp = [str(i) for i in df1["File"]]
axes.plot(temp, df1["DFS Runtime (ms)"],"r-",label = "DFS_runtime",linestyle = "dashed",marker = 'o')
axes.plot(temp, df1["SLS Runtime (ms)"],"y-",label = "SLS",linestyle = "dashed",marker = 'o')
axes.legend()
```

```python
df2 =df[df["File"]=="20_5.0_1.0.out"].copy()
df2
```

```python
fig,axes = plt.subplots(figsize = (18,8))
axes.set_xlabel("heuristic steps")
axes.set_ylabel("runtime 20_5.0_1.0.out")
temp = [2,3,4,5]
axes.plot(temp, df2["DFS Runtime (ms)"],"r-",label = "DFS_runtime")
axes.plot(temp, df2["SLS Runtime (ms)"],"y-",label = "SLS_runtime")
axes.set_xticks(temp)
axes.legend()
```

```python
fig,axes = plt.subplots(figsize = (18,8))
axes.set_xlabel("heuristic steps")
axes.set_ylabel("Solution Quality 20_5.0_1.0.out")
temp = [2,3,4,5]
axes.plot(temp, df2["DFS Solution Quality"],"r-",label = "DFS Solution Quality")
axes.plot(temp, df2["SLS Solution Quality"],"y-",label = "SLS Solution Quality")
axes.set_xticks(temp)
axes.legend()
```
